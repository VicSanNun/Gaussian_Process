```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Defina os parmetros da distribuio gaussiana
mean = [0, 0]  # Mdia das duas dimenses
cov = [[1, 0.5], [0.5, 1]]  # Matriz de covarincia

# Crie uma grade de pontos para avaliar a densidade de probabilidade
x, y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
pos = np.dstack((x, y))
rv = multivariate_normal(mean, cov)

# Calcule a densidade de probabilidade para cada ponto da grade
z = rv.pdf(pos)

# Crie um grfico 3D para visualizar a distribuio
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('Distribuio Gaussiana Bidimensional')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Densidade de Probabilidade')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Defina os parmetros da distribuio normal
mean = 0  # Mdia
std_dev = 1  # Desvio padro

# Gere amostras da distribuio normal
num_samples = 90000
samples = np.random.normal(mean, std_dev, num_samples)

# Crie um histograma para visualizar a distribuio
plt.hist(samples, bins=30, density=True, alpha=0.3, color='b')

# Crie uma curva de densidade de probabilidade terica
x = np.linspace(-4, 4, 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x - mean)**2 / (2 * std_dev**2))
plt.plot(x, pdf, 'r-', lw=2)

plt.title('Distribuio Normal Unidimensional')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(False)
plt.savefig('normal.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Defina os parmetros da distribuio normal
mean = 0  # Mdia
std_dev = 1  # Desvio padro

# Gere amostras da distribuio normal
num_samples = 100000
samples = np.random.normal(mean, std_dev, num_samples)

# Crie um histograma para visualizar a distribuio
plt.hist(samples, bins=30, density=True, alpha=0.2, color='grey')

# Crie uma curva de densidade de probabilidade terica
x = np.linspace(-4, 4, 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x - mean)**2 / (2 * std_dev**2))
plt.plot(x, pdf, 'r-', lw=2, alpha=0.5)

# Desenhe uma linha vertical para a mdia
plt.axvline(mean, color='green', linestyle='--', linewidth=1, label=f'Mdia: {mean}')

# Desenhe uma linha horizontal para o desvio padro
#plt.axhline(0, color='blue', linestyle='-.', linewidth=2, label=f'Desvio Padro: {std_dev}')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.legend()
plt.title('Distribuio Normal Unidimensional')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(False)

plt.savefig('normal.png')

plt.show()
```
